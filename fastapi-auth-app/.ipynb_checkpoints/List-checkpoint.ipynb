```python
# Swap Two Elements
```

```python
num = [1, 2, 3, 4]
num[0], num[1] = num[1], num[0]
print(num)  # Output: [2, 1, 3, 4]
```

```python
# Reverse a List
```

```python
num = [1, 2, 3, 4]

# In-place
num.reverse()
print(num)  # Output: [4, 3, 2, 1]

# Using slicing
reversed_list = num[::-1]
print(reversed_list) # Output: [2, 1, 3, 4]
```

```python
# Sort a List
```

```python
num = [4, 2, 1, 3]

num.sort()
print(num)  # Output: [1, 2, 3, 4]

# Descending
num.sort(reverse=True)
```

```python
# List Comprehension
```

```python
# Squares of numbers
squares = [x**2 for x in range(5)]
print(squares)  # Output: [0, 1, 4, 9, 16]

# Conditional list comprehension
even_squares = [x**2 for x in range(5) if x % 2 == 0]
print(even_squares)  # Output: [0, 4, 16]
```

```python
# Flatten a Nested List
```

```python
nested_list = [[1,2],[3,4],[5,6]]
flat = [x for sublist in nested_list for x in sublist]
print(flat)  # Output: [1, 2, 3, 4, 5, 6]
```

```python
# Find Index of Element
```

```python
num = [10, 20, 30]
index = num.index(20)
print(index)  # Output: 1
```

```python
# Merge Two Lists
```

```python
a = [1, 2, 3]
b = [4, 5, 6]

merged = a + b
# or using unpacking
merged = [*a, *b]
print(merged)  # Output: [1, 2, 3, 4, 5, 6]
```

```python
# Use zip() to Iterate Multiple Lists
```

```python
a = [1,2,3]
b = ['x','y','z']
for num, char in zip(a,b):
    print(num, char)
```

```python
# List Slicing Tricks
```

```python
num = [0,1,2,3,4,5]
print(num[::2])  # Every 2nd element: [0, 2, 4]
print(num[::-1])  # Reverse list: [5, 4, 3, 2, 1]
```

```python
# Insert and Extend
```

```python
num = [1,2,3]
num.insert(1, 10)  # Insert 10 at index 1
num.extend([4,5])  # Add multiple elements
print(num)  # Output: [1, 10, 2, 3, 4, 5]
```

```python
# Use a List as a Stack
```

```python
stack = []
stack.append(1)  # Push
stack.append(2)
print(stack.pop())  # Output: 2
```

```python
# Use a List as a Queue
```

```python
from collections import deque
queue = deque([1,2,3])
queue.append(4)  # Enqueue
print(queue.popleft())  # Output: 1
```

```python
# Advanced Filtering
```

```python
#del list
num = [1,2,3,4,5]
filtered = list(filter(lambda x: x % 2 == 0, num))
print(filtered)  # Output: [2, 4]
```

```python
# Remove Duplicates While Preserving Order
```

```python
num = [1,2,2,3,1]
unique = list(dict.fromkeys(num))
print(unique)  # Output: [1, 2, 3]
```

```python
# Combine Filtering and Mapping
```

```python
num = [1, 2, 3, 4, 5]
squared_evens = [x**2 for x in num if x % 2 == 0]
print(squared_evens)  # Output: [4, 16]
```

```python
# Iterate with Enumerate
```

```python
num = [10, 20, 30]
for i, val in enumerate(num):
    print(i, val)
```
